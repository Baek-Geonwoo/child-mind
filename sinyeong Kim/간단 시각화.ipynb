```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Markdown
import gc
from pandas.api.indexers import FixedForwardWindowIndexer

# series, th, window = "038441c925bb", 50, 8640
def get_train_series(series, th, window, check_point=5*60*12):
  train_series = pd.read_csv(f"/content/drive/MyDrive/ai경진대회/Child Mind/DataPrepare_720/train_csvs/{series}.csv")
  train_events = pd.read_csv("/content/drive/MyDrive/ai경진대회/Child Mind/train_events.csv").query('series_id == @series')

  base = (train_events.loc[train_events['event'].isin(['onset', 'wakeup']), ['event', 'step']])
  sample_onset = train_events.loc[train_events['event'] == 'onset', 'step'].dropna()
  sample_wakeup = train_events.loc[train_events['event'] == 'wakeup', 'step'].dropna()
  print("event 갯수 체크", len(sample_onset), len(sample_wakeup), len(sample_onset) == len(sample_wakeup))
  if len(sample_onset) == 0 or len(sample_wakeup) == 0:
      print("\n\n\n-------", series, "-------\n\n\n")
      print(train_events.loc[train_events['event'] == 'onset', 'step'])
      print(train_events.loc[train_events['event'] == 'wakeup', 'step'])
  elif len(sample_onset) != len(sample_wakeup):
      print("\n\n\n-------", series, "-------\n\n\n")
      print(train_events.loc[train_events['event'] == 'onset', 'step'])
      print(train_events.loc[train_events['event'] == 'wakeup', 'step'])
  # cleaning etc.
  print("train_events 체크", print(len(train_events), len(train_events.dropna())))
  train_events = train_events.dropna()
  train_events["step"]  = train_events["step"].astype("int")
  train_events["awake"] = train_events["event"].replace({"onset":1,"wakeup":0})

  train = pd.merge(train_series, train_events[['step','awake']], on='step', how='left')

  # train.info()

  # print(base, type(base))
  #인덱스 리셋해줘야 첫번째를 0으로 인식가능
  base.reset_index(drop=True, inplace=True)
  # print(base, type(base), len(base))
  for index, row in base.iterrows(): #<-방향으로 nan이 아니면 채움
    # print(index, row['step'], row['event'])
    if index == 0 and not np.isnan(row['step']):
      start_step = 0
      end_step = int(row['step'])
      # print(index, start_step, end_step-1)
      train.loc[start_step:end_step-1, 'awake'] = np.where(row['event'] == 'onset', 1, 0)
      # print(train.iloc[0])
      # print(train.iloc[7019])
      # print(train.iloc[7020])
      # print(train.iloc[7021])
    elif index != 0 and (not np.isnan(row['step']) and not np.isnan(base.at[index-1, 'step'])):
        start_step = int(base.at[index-1, 'step'])
        end_step = int(row['step'])
        # print(index, start_step, end_step-1)
        train.loc[start_step:end_step-1, 'awake'] = np.where(row['event'] == 'onset', 1, 0)
      # print(train.iloc[7000])
      # print(train.iloc[10931])
      # print(train.iloc[10932])
      # print(train.iloc[10933])

    if index == len(base)-1 and not np.isnan(row['step']): #마지막이 nan이 아니면 ->로 다 채움, 따라서 +1부터 처리!
      start_step = int(row['step'])
      # print("final check")
      # print(train.iloc[start_step])
      print("last num", start_step, start_step+check_point, train['step'].iloc[-1])
      train.loc[start_step+1:min(start_step+check_point, train['step'].iloc[-1]), 'awake'] = np.where(row['event'] == 'onset', 0, 1)
      # print(train.iloc[start_step])
      # print(train.iloc[start_step+1])

  train['awake'].fillna(2, inplace=True)

  # train["awake"] = train["awake"].bfill(axis ='rows')

    #rolling을 통하여 특정 값 이하의 값을 2로 처리하고 나머지를 1로 처리함
  #     train['anglez_backward_max'] = train_series['anglez'].rolling(window=window, min_periods=window).max()
  #     train['anglez_forward_max'] = train_series['anglez'].rolling(FixedForwardWindowIndexer(window_size=window), min_periods=window).max()
  #     cond = (train.anglez_backward_max<th)|(train.anglez_forward_max<th)
  #     train.loc[cond, 'awake'] = 2
  #     # Result: the last event is always a "wakeup"
  #     train['awake'] = train['awake'].fillna(1) # awake
  #     train["awake"] = train["awake"].astype("int")
  return train, sample_onset, sample_wakeup
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Markdown
import gc
from pandas.api.indexers import FixedForwardWindowIndexer

# series, th, window = "038441c925bb", 50, 8640
def get_train_series(series, th, window, check_point=5*60*12):
  train_series = pd.read_csv(f"/content/drive/MyDrive/ai경진대회/Child Mind/DataPrepare_720/train_csvs/{series}.csv")
  train_events = pd.read_csv("/content/drive/MyDrive/ai경진대회/Child Mind/train_events.csv").query('series_id == @series')

  base = (train_events.loc[train_events['event'].isin(['onset', 'wakeup']), ['event', 'step']])
  sample_onset = train_events.loc[train_events['event'] == 'onset', 'step'].dropna()
  sample_wakeup = train_events.loc[train_events['event'] == 'wakeup', 'step'].dropna()
  print("event 갯수 체크", len(sample_onset), len(sample_wakeup), len(sample_onset) == len(sample_wakeup))
  if len(sample_onset) == 0 or len(sample_wakeup) == 0:
      print("\n\n\n-------", series, "-------\n\n\n")
      print(train_events.loc[train_events['event'] == 'onset', 'step'])
      print(train_events.loc[train_events['event'] == 'wakeup', 'step'])
  elif len(sample_onset) != len(sample_wakeup):
      print("\n\n\n-------", series, "-------\n\n\n")
      print(train_events.loc[train_events['event'] == 'onset', 'step'])
      print(train_events.loc[train_events['event'] == 'wakeup', 'step'])
  # cleaning etc.
  print("train_events 체크", print(len(train_events), len(train_events.dropna())))
  train_events = train_events.dropna()
  train_events["step"]  = train_events["step"].astype("int")
  train_events["awake"] = train_events["event"].replace({"onset":1,"wakeup":0})

  train = pd.merge(train_series, train_events[['step','awake']], on='step', how='left')

  # train.info()

  # print(base, type(base))
  #인덱스 리셋해줘야 첫번째를 0으로 인식가능
  base.reset_index(drop=True, inplace=True)
  # print(base, type(base), len(base))
  for index, row in base.iterrows(): #<-방향으로 nan이 아니면 채움
    # print(index, row['step'], row['event'])
    if index == 0 and not np.isnan(row['step']):
      start_step = 0
      end_step = int(row['step'])
      # print(index, start_step, end_step-1)
      train.loc[start_step:end_step-1, 'awake'] = np.where(row['event'] == 'onset', 1, 0)
      # print(train.iloc[0])
      # print(train.iloc[7019])
      # print(train.iloc[7020])
      # print(train.iloc[7021])
    elif index != 0 and (not np.isnan(row['step']) and not np.isnan(base.at[index-1, 'step'])):
        start_step = int(base.at[index-1, 'step'])
        end_step = int(row['step'])
        # print(index, start_step, end_step-1)
        train.loc[start_step:end_step-1, 'awake'] = np.where(row['event'] == 'onset', 1, 0)
      # print(train.iloc[7000])
      # print(train.iloc[10931])
      # print(train.iloc[10932])
      # print(train.iloc[10933])

    if index == len(base)-1 and not np.isnan(row['step']): #마지막이 nan이 아니면 ->로 다 채움, 따라서 +1부터 처리!
      start_step = int(row['step'])
      # print("final check")
      # print(train.iloc[start_step])
      print("last num", start_step, start_step+check_point, train['step'].iloc[-1])
      train.loc[start_step+1:min(start_step+check_point, train['step'].iloc[-1]), 'awake'] = np.where(row['event'] == 'onset', 0, 1)
      # print(train.iloc[start_step])
      # print(train.iloc[start_step+1])

  train['awake'].fillna(2, inplace=True)

  # train["awake"] = train["awake"].bfill(axis ='rows')

    #rolling을 통하여 특정 값 이하의 값을 2로 처리하고 나머지를 1로 처리함
  #     train['anglez_backward_max'] = train_series['anglez'].rolling(window=window, min_periods=window).max()
  #     train['anglez_forward_max'] = train_series['anglez'].rolling(FixedForwardWindowIndexer(window_size=window), min_periods=window).max()
  #     cond = (train.anglez_backward_max<th)|(train.anglez_forward_max<th)
  #     train.loc[cond, 'awake'] = 2
  #     # Result: the last event is always a "wakeup"
  #     train['awake'] = train['awake'].fillna(1) # awake
  #     train["awake"] = train["awake"].astype("int")
  return train, sample_onset, sample_wakeup
```

```python

# 'enmo' 열을 미분하여 '속도' 열 생성
train['enmo_diff'] = train['enmo'].diff()  # 현재 값과 이전 값의 차이를 계산

# 첫 번째 행의 '속도' 값은 NaN일 수 있어서, 이를 0으로 채우거나 원하는 방식으로 처리 가능
train['enmo_diff'].fillna(0, inplace=True)

# 시각화
fig, ax = plt.subplots(figsize=(20, 3))
sns.lineplot(data=train[10000:14000], x="step", y="enmo", hue="awake", linewidth=0.5, palette={0: 'black', 1: 'green'}, label="Original Data")
fig, ax = plt.subplots(figsize=(20, 3))
sns.lineplot(data=train[10000:14000], x="step", y="enmo_diff", hue="awake", palette={0: 'red', 1: 'yellow'}, label="diff Data")
plt.show()
```
